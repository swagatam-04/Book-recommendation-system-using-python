```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=8500,stop_words='english')
```

```python
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return" ".join(y)
```

```python
new_dataframe['tags']= new_dataframe['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity (vectors)
```

```python
cosine_similarity (vectors).shape
```

```python
def recommend (book):
    book_index=new_dataframe[new_dataframe['Title']==book].index[0]
    distances=cosine_similarity (vectors)[book_index]
    names_of_books=sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in names_of_books:
        print(new_dataframe.iloc[i[0]].Title)
```

```python
recommend('The Silmarillion')
```

```python
recommend('West with Giraffes: A Novel')
```
